## Observations and Insights 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st

mouse_data=pd.read_csv("data/Mouse_metadata.csv")
study_results=pd.read_csv("data/Study_results.csv")

combined_data=pd.merge(mouse_data,study_results,how="right",on="Mouse ID")
combined_data.head()
```

```python
# Checking the number of mice.
combined_new=combined_data["Mouse ID"].unique()
print(len(combined_new))
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mouse=combined_data.loc[combined_data.duplicated(subset=["Mouse ID","Timepoint"]),"Mouse ID"]
duplicate_mouse
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_data=combined_data.loc[combined_data["Mouse ID"]=="g989"]
duplicate_data
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_study=combined_data[combined_data["Mouse ID"].isin(duplicate_mouse)==False]
clean_study
```

```python
# Checking the number of mice in the clean DataFrame.
len(clean_study["Mouse ID"].unique())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
means=clean_study.groupby("Drug Regimen").mean()["Tumor Volume (mm3)"]
median=clean_study.groupby("Drug Regimen").median()["Tumor Volume (mm3)"]
variance=clean_study.groupby("Drug Regimen").var()["Tumor Volume (mm3)"]
standardDe=clean_study.groupby("Drug Regimen").std()["Tumor Volume (mm3)"]
sems=clean_study.groupby("Drug Regimen").sem()["Tumor Volume (mm3)"]

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.

summaryDF=pd.DataFrame({"Mean":means,
                       "Median":median,
                       "Variance":variance,
                       "Standard Deviation":standardDe,
                       "SEM":sems})
summaryDF
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line
summary_new=clean_study.groupby("Drug Regimen").agg({"Tumor Volume (mm3)":["mean","median","var","std","sem"]})
summary_new
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  

# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
last_timepoint=clean_study.groupby("Mouse ID")["Timepoint"].max()
last_timepoint=last_timepoint.reset_index()
last_timepoint.head()

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
clean_merge=last_timepoint.merge(clean_study,on=["Mouse ID","Timepoint"],how="left")

clean_merge.head()
```

```python
print(clean_merge["Drug Regimen"].unique())
```

```python
# Put treatments into a list for for loop (and later for plot labels)
drug_list=["Capomulin", "Ramicane", "Infubinol","Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumorV=[]

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for v in drug_list:
    
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    tvolumes=clean_merge.loc[clean_merge["Drug Regimen"]==v,"Tumor Volume (mm3)"]
    
    # add subset 
    tumorV.append(tvolumes)
    
    # Determine outliers using upper and lower bounds
    quartiles=tvolumes.quantile([0.25,0.5,0.75])
    lowerq=quartiles[0.25]
    upperq=quartiles[0.75]
    iqr=upperq-lowerq
    lowerbound=lowerq-(1.5*iqr)
    upperbound=upperq+(1.5*iqr)
    outliers=tvolumes.loc[(tvolumes<lowerbound)|(tvolumes>upperbound)]
    print(f"{v}:\n{outliers}\n")
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
orange_out=dict(markerfacecolor="orange",markersize=20)
plt.boxplot(tumorV,labels=drug_list,flierprops=orange_out)
plt.xlabel("Drug")
plt.ylabel("Final Tumor Volume")
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin
Cap_data=clean_study.loc[clean_study["Drug Regimen"]=="Capomulin"]
Cap_data
```

```python
Cap_mouse=Cap_data.loc[Cap_data["Mouse ID"]=="r554"]
plt.plot(Cap_mouse["Timepoint"],Cap_mouse["Tumor Volume (mm3)"])
plt.xlabel("Timepoint")
plt.ylabel("Tumor Volume")
plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
ave_cap=Cap_data.groupby("Mouse ID").mean()
ave_cap
```

```python
plt.scatter(ave_cap["Weight (g)"],ave_cap["Tumor Volume (mm3)"])
plt.xlabel("Weight")
plt.ylabel("Tumor Volume")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
correlation=round(st.pearsonr(ave_cap["Weight (g)"],ave_cap["Tumor Volume (mm3)"])[0],2)
correlation
```

```python
regression=st.linregress(ave_cap["Weight (g)"],ave_cap["Tumor Volume (mm3)"])
regression
```

```python
yvalue=ave_cap["Weight (g)"]*regression[0]+regression[1]
plt.plot(ave_cap["Weight (g)"],yvalue,color="red")
plt.show()
```
